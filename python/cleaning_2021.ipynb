# 2021 CLEANING
- loaded in 2021
- select the columns
- rename the columns
- clean the columns
- saved the clean 2021

```python
# test print some of the columns
df_2021 = pd.read_csv("../data/raw/nyts2021.csv", low_memory=False)
df_2021.shape
```

```python
# get only columns needed for analysis
cols_2021 = [
    "QN1", "QN2", "QN3",
    "QN133", "QN134",
    "QN135A", "QN135B", "QN135C", "QN135D",
    "QN135E", "QN135F", "QN135G", "QN135H",
    "QN136", "QN137",
    "QN138A", "QN138B", "QN138C", "QN138D", "QN138E", "QN138F", "QN138G",
    "QN9"
]

# save as subset
df_2021_small = df_2021[cols_2021].copy()
df_2021_small.head()
```

```python
# rename the columns
df_2021_small = df_2021_small.rename(columns={
    "QN1": "age",
    "QN2": "sex",
    "QN3": "grade",
    "QN133": "sm_usage",
    "QN134": "sm_freq",
    "QN135A": "platform_fb", #facebook
    "QN135B": "platform_insta", #instagram
    "QN135C": "platform_sc", # snapchat
    "QN135D": "platform_tt", #tiktok
    "QN135E": "platform_twitter", #twitter
    "QN135F": "platform_reddit", #reddit
    "QN135G": "platform_yt", #youtube
    "QN135H": "platform_other", #other
    "QN136": "post_freq",
    "QN137": "interact_freq",
    "QN138A": "creator_real_life",
    "QN138B": "creator_online_friends",
    "QN138C": "creator_influencers",
    "QN138D": "creator_brands_sellers",
    "QN138E": "creator_news",
    "QN138F": "creator_public_health",
    "QN138G": "creator_other",
    "QN9": "usage" #target variable
})

# see the renamed dataset
df_2021_small.head()
```

```python
#look at values prior to cleaning
for col in df_2021_small.columns:
    print(f"\n{col}")
    print(df_2021_small[col].value_counts(dropna=False).sort_index().head(15))

# make sure all columns are numeric
df_2021_small = df_2021_small.apply(pd.to_numeric, errors="coerce")

# check the missing values before cleaning
df_2021_small.isna().sum()
```

```python
# fill in the multiselect binary columns: 1 = selected, 0 = not selected
binary_cols = [
    "platform_fb","platform_insta","platform_sc","platform_tt",
    "platform_twitter","platform_reddit","platform_yt","platform_other",
    "creator_real_life","creator_online_friends","creator_influencers",
    "creator_brands_sellers","creator_news","creator_public_health","creator_other"
]

df_2021_small[binary_cols] = df_2021_small[binary_cols].fillna(0)

# drop rows where the target variable is missing
df_2021_small = df_2021_small[df_2021_small["usage"].notna()].copy()
```

```python
# make the target variable binary: 1 = current ecig user; 0 = no current use
df_2021_small["ecig_current"] = (df_2021_small["usage"] > 0).astype(int)

# median-impute remaining ordinal / demog columns
ordinal_cols = [
    "age", "sex", "grade",
    "sm_usage", "sm_freq",
    "post_freq", "interact_freq"
]

df_2021_small[ordinal_cols] = df_2021_small[ordinal_cols].fillna(
    df_2021_small[ordinal_cols].median()
)

# add in a year column# add year column
df_2021_small["year"] = 2021
df_2021_small["platform_twitch"] = 0 # twitch was not included in 2021 survey so we are adding it here
df_2021_small.isna().sum() # extra missing value check
```

```python
# target balance check
df_2021_small["ecig_current"].value_counts(normalize=True)
```

```python
# preview final cleaned dataset
df_2021_small.head()

# save cleaned 2021 file
df_2021_small.to_csv("../data/cleaned/nyts2021_clean.csv", index=False)
```

```python
# load 2023 raw data
df_2023 = pd.read_csv("../data/raw/nyts2023.csv", low_memory=False)
# test check
print(df_2023.shape)
```

# Merge Cleaned NYTS Datasets (2021-2023)

```python
# load in cleaned datasets
df_2021_clean = pd.read_csv("../data/cleaned/nyts2021_clean.csv")
df_2022_clean = pd.read_csv("../data/cleaned/nyts2022_clean.csv")
df_2023_clean = pd.read_csv("../data/cleaned/nyts2023_clean.csv")

# add in order so it's same across all datasets
cols_order = [
    "age","sex","grade",
    "sm_usage","sm_freq",
    "platform_fb","platform_insta","platform_sc","platform_tt",
    "platform_twitch",
    "platform_twitter","platform_reddit","platform_yt","platform_other",
    "post_freq","interact_freq",
    "creator_real_life","creator_online_friends","creator_influencers",
    "creator_brands_sellers","creator_news","creator_public_health","creator_other",
    "usage","ecig_current","year"
]

df_2021_clean = df_2021_clean[cols_order]
df_2022_clean = df_2022_clean[cols_order]
df_2023_clean = df_2023_clean[cols_order]

# double check shape
print(df_2021_clean.shape)
print(df_2022_clean.shape)
print(df_2023_clean.shape)
```

```python
# double checking column names
print(df_2021_clean.columns)
print(df_2022_clean.columns)
print(df_2023_clean.columns)
```

```python
# run the merge
df_all = pd.concat(
    [df_2021_clean, df_2022_clean, df_2023_clean],
    ignore_index=True
)

df_all.shape # check shape (3590 +5720 +3278 = 12588)
```

```python
df_all["year"].value_counts() # check year distribution
```

```python
df_all.isna().sum().sum() # missing value check
```

```python
# save clean dataset concatinating 2021-2023!!!!
df_all.to_csv("../data/cleaned/nyts_2021_2023_clean.csv", index=False)
```
